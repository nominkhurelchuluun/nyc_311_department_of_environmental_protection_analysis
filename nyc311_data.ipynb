```python
import requests
import pandas as pd

url = "https://data.cityofnewyork.us/resource/erm2-nwe9.json"

params = {
    "$where": "agency = 'DEP' AND created_date >= '2026-01-01' AND created_date <= '2026-06-17'",
    "$limit": 50000,
    "$offset": 0,
    "$order": "created_date ASC"
}

all_records = []

while True:
    r = requests.get(url, params=params)
    batch = r.json()
    if not batch:
        break
    all_records.extend(batch)
    print(f"Fetched {len(all_records)} rows...")
    if len(batch) < 50000:
        break
    params["$offset"] += 50000

df = pd.DataFrame(all_records)
df.to_csv("nyc_311_DEP_2026.csv", index=False)
print(f"Done — {len(df)} rows saved.")
```
